```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 读入数据
train = pd.read_csv("train_trans.csv")
# 查看表格头
print(train.head())
```

```python
# 计算相关性
d = train.corr()
display(d)
```

```python
# sns.set(style="white",font_scale=0.8)#设置主题，文本大小
# 画热力图
plt.subplots(figsize = (12,12))
sns.heatmap(d,annot = True,vmax = 1,square = True,cmap = "Oranges")
plt.show()
```

```python
# 配置颜色
colors = {0: '#F8DA63', 1: '#BCC957', 2: '#81B655', 3:'#45A058', 4:'#00885C', 5:'#00705E', 6:'#005246'}
sns.jointplot(data=train, x="Height", y="Weight", hue="NObeyesdad",height=6,palette=colors,marker='^')
```
